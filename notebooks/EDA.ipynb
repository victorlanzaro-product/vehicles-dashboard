```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../vehicles.csv')
df.head()
```

```python
fig = px.histogram(df, x='odometer', title='Distribuição do Odômetro')
fig.show()
```

```python
fig = px.scatter(df, x='odometer', y='price', title='Odômetro vs Preço')
fig.show()
```
